```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Dictionary to store MSE values for each feature
mse_results = {}

# Evaluate each feature individually
for column in x_train.columns:
    model = LinearRegression()
    model.fit(x_train[[column]], y_train)  # Train on single feature
    y_pred = model.predict(x_train[[column]])  # Predict
    mse = mean_squared_error(y_train, y_pred)  # Compute MSE
    mse_results[column] = mse  # Store result

# Convert results to DataFrame and sort by MSE (ascending)
mse_df = pd.DataFrame.from_dict(mse_results, orient='index', columns=['MSE']).sort_values(by='MSE')

# Plot the results
plt.figure(figsize=(12, 6))
sns.barplot(x=mse_df.index, y=mse_df['MSE'], palette='coolwarm')

# Customize plot
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.ylabel('Mean Squared Error (MSE)', fontsize=14)
plt.xlabel('Features', fontsize=14)
plt.title('Feature Importance Based on Individual Linear Regression MSE', fontsize=16, fontweight='bold')
plt.grid(axis='y', alpha=0.3)

# Show plot
plt.show()

# Display sorted results
print(mse_df)

```

```python
# without outliers
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# Function to remove outliers using IQR method
def remove_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]

# Store MSE results before and after outlier removal
mse_results = {"With Outliers": {}, "Without Outliers": {}}

# Evaluate each feature individually
for column in x_train.columns:
    # Train model with outliers
    model = LinearRegression()
    model.fit(x_train[[column]], y_train)
    y_pred = model.predict(x_train[[column]])
    mse_results["With Outliers"][column] = mean_squared_error(y_train, y_pred)

    # Remove outliers
    df_clean = remove_outliers(pd.concat([x_train[[column]], y_train], axis=1), column)
    
    # Train model without outliers
    if not df_clean.empty:  # Ensure data exists after removing outliers
        model.fit(df_clean[[column]], df_clean[y_train.name])
        y_pred_clean = model.predict(df_clean[[column]])
        mse_results["Without Outliers"][column] = mean_squared_error(df_clean[y_train.name], y_pred_clean)
    else:
        mse_results["Without Outliers"][column] = np.nan  # If all data is removed

# Convert results to DataFrame
mse_df = pd.DataFrame(mse_results)

# Plot comparison
plt.figure(figsize=(14, 6))
mse_df.plot(kind="bar", figsize=(14, 6), color=["red", "blue"], alpha=0.7, edgecolor='black')

# Customize plot
plt.xticks(rotation=45, ha="right", fontsize=12)
plt.ylabel("Mean Squared Error (MSE)", fontsize=14)
plt.xlabel("Features", fontsize=14)
plt.title("Feature Importance Before & After Outlier Removal", fontsize=16, fontweight="bold")
plt.legend(["With Outliers", "Without Outliers"])
plt.grid(axis="y", alpha=0.3)

# Show plot
plt.show()

# Display numerical comparison
print(mse_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score

# Загружаем датасет


file_path_ext = "../../data/BodyFat - Extended.csv"
df = pd.read_csv(file_path_ext)


# Проверяем данные
print(df.head())
print(df.info())

# Разделяем мужчин и женщин
men = df[df['Sex'] == 'M'].copy()
women = df[df['Sex'] == 'F'].copy()

# Убираем ненужные столбцы (ID, имя и т.п., если есть)
men = men.select_dtypes(include=[np.number]).dropna()
women = women.select_dtypes(include=[np.number]).dropna()

# Целевая переменная (процент жира)
target = 'BodyFat'

# Функция для обучения линейной регрессии и расчета важности признаков
def train_and_analyze(df, gender):
    X = df.drop(columns=[target])
    y = df[target]
    
    # Стандартизация
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Обучаем модель
    model = LinearRegression()
    model.fit(X_scaled, y)
    
    # Предсказание
    y_pred = model.predict(X_scaled)
    
    # Оцениваем модель
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    print(f"{gender} - MAE: {mae:.2f}, R2: {r2:.2f}")
    
    # Важность признаков
    feature_importance = pd.Series(np.abs(model.coef_), index=X.columns).sort_values(ascending=False)
    
    # Визуализация
    plt.figure(figsize=(10, 5))
    sns.barplot(x=feature_importance.values, y=feature_importance.index)
    plt.xlabel("Feature Importance (Absolute Coefficients)")
    plt.ylabel("Features")
    plt.title(f"Feature Importance - {gender}")
    plt.show()
    
    return feature_importance

# Запуск анализа
print("Analyzing men...")
men_importance = train_and_analyze(men, "Men")
print("Analyzing women...")
women_importance = train_and_analyze(women, "Women")
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, r2_score

# Load dataset
file_path_ext = "../../data/BodyFat - Extended.csv"
df = pd.read_csv(file_path_ext)

# Filter men and women
men = df[df['Sex'] == 'M'].copy()
women = df[df['Sex'] == 'F'].copy()

# Keep only numerical columns and drop NaN values
men = men.select_dtypes(include=[np.number]).dropna()
women = women.select_dtypes(include=[np.number]).dropna()

# Target variable
target = 'BodyFat'

# Function to train linear regression and analyze feature importance
def train_and_analyze(df, gender):
    X = df.drop(columns=[target])
    y = df[target]
    
    # Standardization
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    
    # Train model
    model = LinearRegression()
    model.fit(X_scaled, y)
    
    # Prediction
    y_pred = model.predict(X_scaled)
    
    # Model evaluation
    mae = mean_absolute_error(y, y_pred)
    r2 = r2_score(y, y_pred)
    print(f"{gender} - MAE: {mae:.2f}, R2: {r2:.2f}")
    
    # Feature importance
    feature_importance = pd.Series(np.abs(model.coef_), index=X.columns).sort_values(ascending=False)
    
    return feature_importance

# Run analysis
print("Analyzing men...")
men_importance = train_and_analyze(men, "Men")
print("Analyzing women...")
women_importance = train_and_analyze(women, "Women")

# Combine feature importance for comparison
importance_df = pd.DataFrame({'Men': men_importance, 'Women': women_importance}).dropna()

# Correlation between men and women coefficients
correlation = importance_df.corr().iloc[0, 1]

# Visualization
plt.figure(figsize=(12, 6))
sns.heatmap(importance_df.corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=2)
plt.title(f"Correlation of Feature Importance between Men and Women: {correlation:.2f}")
plt.show()

# Display importance comparison
importance_df.plot(kind='bar', figsize=(14, 6), alpha=0.75)
plt.xlabel("Features")
plt.ylabel("Absolute Coefficients")
plt.title("Comparison of Feature Importance Between Men and Women")
plt.xticks(rotation=45)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.legend(["Men", "Women"])
plt.show()
```
